```python
#import TRUMP TWEET DATA
trump_tweets = "Resources/trumptweets.csv"
TRUMP_df = pd.read_csv(trump_tweets)
TRUMP_df.head()
```

```python
New_TRUMP_df = TRUMP_df[['created_at', 'text', 'mentions_screen_name','hashtags']].copy()
New_TRUMP_df = New_TRUMP_df.rename(columns={'created_at': 'Date'})


New_TRUMP_df.head()
```

```python
New_TRUMP_df.Date=New_TRUMP_df.Date.str[:10]
New_TRUMP_df.head()
```

```python
#Import DOW DATA
DOW_Data = "Resources/DJI.csv"
DOW_df = pd.read_csv(DOW_Data)
DOW_df.rename(columns={'Close': 'DOW_Close'}, inplace=True)
DOW_df = DOW_df[['Date', 'DOW_Close']].copy()
DOW_df.head()
```

```python
#Import S&P Data
SP_Data = "Resources/GSPC.csv"
SP_df = pd.read_csv(SP_Data)
SP_df.rename(columns={'Close': 'S&P_Close'}, inplace=True)
SP_df = SP_df[['Date', 'S&P_Close']].copy()
SP_df.head()
```

```python
#Import Nasdaq Data & rename close column
Nasdaq_Data = "Resources/IXIC.csv"
Nasdaq_df = pd.read_csv(Nasdaq_Data)
Nasdaq_df.rename(columns={'Close': 'Nasdaq_Close'}, inplace=True)
Nasdaq_df = Nasdaq_df[['Date', 'Nasdaq_Close']].copy()
Nasdaq_df.head()
```

```python
#Connect to Local Database
rds_connection_string = "root:password@localhost/ETL?charset=utf8mb4"
engine = create_engine(f'mysql+pymysql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python

Phase1=(pd.merge(New_TRUMP_df,DOW_df,on='Date', how='outer').fillna({'comp_id':''}))
Phase2=(pd.merge(Phase1,SP_df,on='Date', how='outer').fillna({'comp_id':''}))
Phase3=(pd.merge(Phase2,Nasdaq_df,on='Date', how='outer').fillna({'comp_id':''}))
Phase3.head()
```
